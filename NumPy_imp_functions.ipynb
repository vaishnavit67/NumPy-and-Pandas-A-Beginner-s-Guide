```python
# np.array_equal() - to check if two arrays match exactly
p = np.array([1, 2, 3])
q = np.array([1, 2, 3])
r = np.array([1, 2, 4])

print(np.array_equal(p, q))
print(np.array_equal(p, r))
```

```python
# np.allclose() - checks if two arrays are practically equal
# within a tiny, acceptable margin of error (tolerance)

a1 = np.array([1.000000001])
b1 = np.array([1.000000000])
print("Case 1:", np.allclose(a1, b1, rtol=1e-05, atol=1e-08))

a2 = np.array([100000.1])
b2 = np.array([100000.0])
print("Case 2:", np.allclose(a2, b2, rtol=1e-05, atol=1e-08))

a3 = np.array([100000.5])
b3 = np.array([100000.0])
print("Case 3:", np.allclose(a3, b3, rtol=1e-06, atol=1e-08))
```
