## Creating NumPy Arrays

```python
# np.array - to create an array
import numpy as np

a = np.array([1,2,3])
print(type(a))
print(a)
```

```python
# 2D and 3D arrays
b = np.array([[1,2,3],[4,5,6]])
print(b)
print(type(b))

c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

```python
# to create an array of a specific data type, float, bool using dtype(datatype)
np.array([1,2,3], dtype=float)
np.array([1,0,1], dtype=bool)
np.array([1,2,3], dtype=complex)
```

```python
# arange function
np.arange(1,11)
np.arange(1,11,2)
```

```python
# reshape function
# to change the shape(dimensions) of an array without changing its actual data
# works with both normal arrays and arange
np.arange(1,16).reshape(5,3)
```

```python
# np.ones, np.zeros, np.random
np.ones((3,4))

np.zeros((2,3))

np.random.random((4,5)) # random class -> random method
```

```python
# np.linspace(range start, range end, no:of items) - with equal distances between them
# generates linearly seperable points
# default - float,
np.linspace(-5,5,10)
np.linspace(-2,6,20, dtype=int)
```

```python
# np.identity - to make an identity matrix
np.identity(3)
```

## Array Attributes


```python
a1 = np.arange(10, dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
a3
```

```python
# ndim - number of dimensions
a3.ndim
```

```python
# shape - gives the no:of items in every dimension
a3.shape
```

```python
# size - gives the no:of items in the array
a3.size
```

```python
#itemsize - size of the datatype used
a1.itemsize
```

```python
# dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

## Changing Datatype

```python
# astype - to change your data type to reduce your memory space consumed
a3.dtype
a3.astype(np.int32)
```

## Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
a1
```

```python
# scaler operations
a2 + 2
a2 - 2
a2 * 2
a2 / 2
a2 % 2
```

```python
# relational operations
a2 > 15
a2 == 15
a2 != 15
```

```python
# vector operations - works only with those having same shape
a1 + a2
a2 - a1
a1 * a2
a1 ** a2
```

## Array Functions

```python
np.random.randint(0,101,size=(3,3))
```

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max, min, sum, prod
# 0 -> col, 1 -> row
np.max(a1)
np.min(a1)
np.sum(a1)
np.prod(a1)
np.min(a1, axis = 0)
np.max(a1, axis=1)
np.sum(a1, axis=0)
```

```python
# mean, median, std, var
np.median(a1)
np.mean(a1, axis=1)
np.var(a1)
np.std(a1)
```

```python
# trigonometric functions
np.sin(a1)
```

```python
# dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(13,25).reshape(4,3)
np.dot(a2,a3)
```

```python
# log and exp
np.log(a1)
np.exp(a1)
```

```python
# round, floor, ceil
np.round(np.random.random((2,3))*100)
np.floor(np.random.random((2,3))*100)
np.ceil(np.random.random((2,3))*100)
```

## Iterating

```python
a1

for i in a1:
  print(i, end = " ")
```

```python
a2
for i in a2:
  print(i)
for i in a2:
  for j in i:
    print(j, end = " ")
  print()
```

```python
a3
for i in a3:
  print(i)
for i in a3:
  for j in i:
    for k in j:
      print(k, end=" ")
    print(end=", ")
  print()
```

```python
# nditer is like getting all elements similar to ravel
for i in np.nditer(a3):
  print(i, end=" ")
```

## Reshaping

```python
# transpose
np.transpose(a2)
a2.T
```

```python
# ravel - can convert any dimension array to 1D array
a2.ravel()
a3.ravel()
```

## Stacking

```python
# Stacking - only works with same dimension arrays
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
```

```python
# horizontal stacking (joins horizontally)
np.hstack((a4,a5))
np.hstack((a4,a5,a4))
```

```python
# vertical stacking (joins vertically)
np.vstack((a4,a5))
```

## Splitting

```python
# horizontal splitting (cuts vertically)
np.hsplit(a4,2)
```

```python
# vertical splitting (cuts horizontally)
np.vsplit(a4,3)
```
