## What is Pandas?
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of Python programming language

### Pandas Series
A pandas series is like a column in a table. It is a 1D array holding data of any type

```python
## Importing Pandas

import numpy as np
import pandas as pd
```

## Making Pandas Series

```python
# Series from lists

# string
country = ['India', 'Pakistan', 'USA', 'Nepal', 'SriLanka']
pd.Series(country)

# integer
runs = [13, 24, 56, 78, 103]
runs_series = pd.Series(runs)
runs_series

# custom indexing
marks = [67, 78, 59, 100]
subjects = ['maths', 'english', 'hindi', 'science']
pd.Series(marks, index = subjects)

# setting a name
marks = pd.Series(marks, index = subjects, name = 'marks of someone')
marks
```

```python
# Series from dictionaries
marks = {
    'maths': 67,
    'english':78,
    'hindi': 59,
    'science': 100
}
marks_series = pd.Series(marks, name='marks of someone')
marks_series
```

```python
## Series attributes

# size
marks_series.size

# dtype
marks_series.dtype

# name
marks_series.name

# is_unique
marks_series.is_unique
pd.Series([1,1,2,3,4,5]).is_unique

# index
marks_series.index
runs_series.index
type(runs_series.index)

# values
marks_series.values
```

## Series indexing

Position vs. Label: The Core Confusion
Pandas has to juggle two ways of looking at a DataFrame or Series:

Position: The 0-based integer location (like standard Python lists).

Label: The actual name/value assigned to that row or column index.

Pandas allows negative indexing unless your index contains integers and doesn't explicitly start from 0.

## Some Important Series Methods

```python
# astype
# between
# clip
# drop_duplicates
# isnull
# dropna
# fillna
# isin
# apply
# copy
```

```python
# astype
import sys
sys.getsizeof(runs)
sys.getsizeof(runs.astype('int16'))
```

```python
# between - both (m,n) are included
runs[runs.between(51,99)].size
```

```python
# clip
subs.clip(100,200)
```

```python
# drop duplicates
temp = pd.Series([1,1,2,2,3,4,4])
temp.drop_duplicates()
temp.drop_duplicates(keep = 'last') # keeps the last index of the value alone and removes rest of the elements
temp.duplicated().sum() # gives how many duplicates exist
```

```python
runs.duplicated().sum()
```

```python
temp = pd.Series([1,2,3,np.nan,5,6,np.nan,8,np.nan,10])
temp
temp.size # gives the count no:of elements in the series
temp.count() # gives the count of no:of non-null elements in the series
```

```python
# isnull - checks if missing values exist
runs.isnull()
runs.isnull().sum()
temp.isnull().sum()
```

```python
# dropna - drops all missing values
temp.dropna()
```

```python
# fillna - fills all null values with some value
temp.fillna(0)
temp.fillna(temp.mean())
```

```python
# isin - checks multiple elements at once
runs[(runs == 49) | (runs == 99)]
runs[runs.isin([49,99])]
```

```python
# apply - allows custom functions to run on the dataset
movies.apply(lambda x:x.split()[0].upper())
subs.apply(lambda x : "good day" if x > subs.mean() else "bad day")
```

```python
# copy
new = runs.head()
new
new[1] = 100
new
runs
# here, head function doesn't give you the copy of the data, rather it gives the original data which changes permanently if you change it
# so use copy
new = runs.head().copy()
new
new[1] = 1
new
runs # doesn't get changed
```
